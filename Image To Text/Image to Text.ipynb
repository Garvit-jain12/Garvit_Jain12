# IMAGE TO TEXT 

## Importing the libraries

```python
import cv2
import os
import pytesseract as tess
tess.pytesseract.tesseract_cmd= 'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

## Getting the file names ( for multiple images only)

```python
filenames = os.listdir('Pik a Pak/Pik a Pak')
```

## Setting the directories

```python
#filenames
directory='Pik a Pak/Pik a Pak/'
output_directory='Pik a Pak/Text/'
```

## Writing the text files in the output directory

```python
for i in filenames:
    name=directory+i
    f=cv2.imread(name)
    text=tess.image_to_string(f)
    L=text.split('\n')
    
    output_name=output_directory+i.split('.')[0]+'.txt'
    file=open(output_name,'w')
    
    for j in L:
        file.write(j)
        if(i!=''):
            file.write('\n')
    file.close()
    
    
    
```
